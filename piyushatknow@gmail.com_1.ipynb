```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns

#import haberman data
cases = pd.read_csv('haberman.csv')

#Classes description
classes = cases['surv_status'].value_counts()
print(classes)

#Here,surv_status : 1 = the patient survived 5 years or longer and surv_status : 2 = the patient died within 5 year
#We can see the data is unbalanced as there is significant range between 225 to 81
```

```python
#Bivariate analysis of various features
#We'll first plot normal scatter plot and then we'll plot the pair-plots

sns.set_style('whitegrid')
sns.FacetGrid(cases,hue='surv_status',size=5,legend_out = True).map(plt.scatter,'age','op_year').add_legend()
plt.show()

'''
Observations : We can see that op_year and age are not good features for Classification
as points overlap each other and there is no scope to put certain conditions for seggregation
'''
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(cases,hue='surv_status',size=5,legend_out = True).map(plt.scatter,'axil_nodes','op_year').add_legend()
plt.show()

'''
Observations : 

1.We can see that axil_nodes and op_year are not good features for Classification
as points overlap each other and there is no scope to put certain conditions for seggregation

W
'''
```

```python
#Now lets try pair plots since we have to observe 4C2 plots for coming to a conclusion

plt.close();
sns.set_style("whitegrid");
sns.pairplot(cases, hue="surv_status", size=3);
plt.show()

'''
Observation : We can observe from pair-plots below is there is no certainity of survival status in any combination of features
Survival status is not dependent or even little biases to any of age/axil_nodes/op_year
We'll have to try univariate analysis
'''
```

```python
#Univariate analysis - PDF , CDF , boxplot , violin plot
#lets see for histogram-distribution fit plot

#hist-dist plot for age

sns.FacetGrid(cases, hue="surv_status", size=6) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();

'''
Observation :
1. Here we can see a man can survive for more than 5 years if he's age is in the range of 30 - 34
1. Here we can see a man cannot survive for more than 5 years if he's age is in the range of 76 - 82
'''
```

```python
#hist-dist plot for axil_nodes

sns.FacetGrid(cases, hue="surv_status", size=6) \
   .map(sns.distplot, "axil_nodes") \
   .add_legend();
plt.show();

'''
Observation :

1.Here we can see, if axillary nodes are between 0 to 2 , chances of surviving for more than 5 years is more 4 times (roughly)
than surviving for less than 5 years

2.1.Here we can see, if axillary nodes are between 2 to 4 , chances of surviving for more less 5 years is more 2 times (roughly)
than surviving for more than 5 years

3. Also we can we observe, that as number of axillary nodes increase than 5 , there is an inverse proportion between surviving
for more than 5 years and surviving for less than 5 years


'''
```

```python
#hist-dist plot for op_year

sns.FacetGrid(cases, hue="surv_status", size=6) \
   .map(sns.distplot, "op_year") \
   .add_legend();
plt.show();

'''
Observation :

Here we can see the operation year doesn't impact much on survival status of the patient

'''
```

```python
#Mean , Variance , Std Deviation

print('For Survival Status = 1')
print("Means")
print('Age : ',np.mean(survival_1["age"]))
print('Axillary Nodes : ',np.mean(survival_1["axil_nodes"]))

print("\nStd-dev");
print('Age : ',np.std(survival_1["age"]))
print('Axillary Nodes : ',np.std(survival_1["axil_nodes"]))

print('\nFor Survival Status = 2')
print("Means")
print('Age : ',np.mean(survival_2["age"]))
print('Axillary Nodes : ',np.mean(survival_2["axil_nodes"]))

print("\nStd-dev");
print('Age : ',np.std(survival_2["age"]))
print('Axillary Nodes : ',np.std(survival_2["axil_nodes"]))
```

```python
#Median, Quantiles, Percentiles, IQR.

from statsmodels import robust

print('For survival status = 1')
print("\nMedians:")
print('Age : ',np.median(survival_1["age"]))
print('Axil Nodes : ',np.median(survival_1["axil_nodes"]))


print("\nQuantiles:")
print('Age : ',np.percentile(survival_1["age"],np.arange(0, 100, 25)))
print('Axil Nodes : ',np.percentile(survival_1["axil_nodes"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print('Age : ',np.percentile(survival_1["age"],90))
print('Axil Nodes : ',np.percentile(survival_1["axil_nodes"],90))


print ("\nMedian Absolute Deviation")
print('Age : ',robust.mad(survival_1["age"]))
print('Axil Nodes : ',robust.mad(survival_1["axil_nodes"]))

print('\nFor survival status = 2')
print("\nMedians:")
print('Age : ',np.median(survival_2["age"]))
print('Axil Nodes : ',np.median(survival_2["axil_nodes"]))


print("\nQuantiles:")
print('Age : ',np.percentile(survival_2["age"],np.arange(0, 100, 25)))
print('Axil Nodes : ',np.percentile(survival_2["axil_nodes"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print('Age : ',np.percentile(survival_2["age"],90))
print('Axil Nodes : ',np.percentile(survival_2["axil_nodes"],90))


print ("\nMedian Absolute Deviation")
print('Age : ',robust.mad(survival_2["age"]))
print('Axil Nodes : ',robust.mad(survival_2["axil_nodes"]))
```
